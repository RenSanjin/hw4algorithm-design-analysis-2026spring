idea:

````python
def sort(arr):
    if len(arr) <= 1:
        return arr
    p = arr[len(arr) // 2]
    left = [x for x in arr if x < p]
    middle = [x for x in arr if x == p]
    right = [x for x in arr if x > p]
    return sort(left) + middle + sort(right)

# algorithm time complexity is: O(n log n)
````

idea:

````python
def max_profit(profits):
    if not profits:
        return 0, -1, -1

    max_profit = float('-inf')
    current_sum = 0

    best_buy_day = 1
    best_sell_day = 1
    temp_buy_day = 1

    for i in range(len(profits)):
        current_profit = profits[i]

        if current_sum < 0:
            current_sum = current_profit
            temp_buy_day = i + 1
        else:
            current_sum += current_profit

        if current_sum > max_profit:
            max_profit = current_sum
            best_buy_day = temp_buy_day
            best_sell_day = i + 1 #

    return max_profit, best_buy_day, best_sell_day


if __name__ == "__main__":
    profits_array = [3, 2, 1, -7, 5, 2, -1, 3, -1]
    max_val, buy, sell = max_profit(profits_array)

    print(f"输入收益数组: {profits_array}")
    print(f"第 {buy} 天买入, 第 {sell} 天卖出, 收益最大：{max_val}")

# algorithm time complexity is: O(n)
````

def find_one_local_minimum(A):
    if not A:
        return None

    n = len(A)
    left, right = 0, n - 1

    while left <= right:
        mid = left + (right - left) // 2

        is_less_than_left = (mid == 0) or (A[mid] < A[mid - 1])
        is_less_than_right = (mid == n - 1) or (A[mid] < A[mid + 1])

        if mid > 0 and A[mid - 1] < A[mid]:

        else:
            left = mid + 1

    return -1

>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]
<div style="text-align:center;width:1000px;"> <img alt="" src="./fig/问题6.jpg"> </div>


idea:

````python
def asteroid_collision_custom(asteroids):
    stack = []

    for ast in asteroids:
        if ast > 0:
            stack.append(ast)
        else:
            alive = True
            while stack and stack[-1] > 0:
                if abs(ast) > stack[-1]:
                    stack.pop()
                elif abs(ast) < stack[-1]:
                    alive = False
                    break
                else:
                    alive = False
                    break
            if alive:
                stack.append(ast)
    return stack

if __name__ == "__main__":
    A1 = [-3, -6, 2, 8, 5, -8, 9, -2, 1]
    print(f"输入: {A1}\n输出: {asteroid_collision_custom(A1)}\n")

    A2 = [23, -8, 9, -3, -7, 9, -23, 22]
    print(f"输入: {A2}\n输出: {asteroid_collision_custom(A2)}\n")

# algorithm time complexity is: O(n)
````

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现
<div style="text-align:center;width:1000px;"> <img alt="" src="./fig/问题7.jpg"> </div>


idea:

````python
def knapsack(items, capacity):

    n = len(items)

    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]
    for i in range(1, n + 1):

        current_weight, current_value = items[i - 1]

        # 内层循环：正序遍历所有的容量状态
        for j in range(1, capacity + 1):

            if j < current_weight:
                dp[i][j] = dp[i - 1][j]

            else:
                dp[i][j] = max(dp[i - 1][j],
                               dp[i - 1][j - current_weight] + current_value)

    return dp[n][capacity]

if __name__ == "__main__":
    marbles = [(5, 10), (4, 40), (6, 30), (3, 50)]
    max_capacity = 9

    max_val = knapsack(marbles, max_capacity)

    print(f"大理石重量与价值: {marbles}")
    print(f"小推车最大承重: {max_capacity}")
    print(f"使用小推车装大理石的最大价值为: {max_val}")
````

输入： root=[3,1,4,null,null,2]， 输出为：[2,1,4,null,null,3]

解释：2 不能在 3 的右子树中，因为 2 < 3 。交换 2 和 3 使二叉搜索树有效。
<div style="text-align:center;width:1000px;"> <img alt="" src="./fig/问题8.jpg"> </div>


idea:

````python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def build_tree(arr):
    if not arr:
        return None
    root = TreeNode(arr[0])
    queue = [root]
    i = 1
    while queue and i < len(arr):
        node = queue.pop(0)
        if arr[i] is not None:
            node.left = TreeNode(arr[i])
            queue.append(node.left)
        i += 1
        if i < len(arr) and arr[i] is not None:
            node.right = TreeNode(arr[i])
            queue.append(node.right)
        i += 1
    return root

def tree_to_list(root):
    if not root:
        return []
    result = []
    queue = [root]
    while queue:
        node = queue.pop(0)
        if node:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)
        else:
            result.append(None)
    while result and result[-1] is None:
        result.pop()
    return result

def inorder_traversal(node, nodes_list):
    if not node:
        return
    inorder_traversal(node.left, nodes_list)
    nodes_list.append(node)
    inorder_traversal(node.right, nodes_list)

def recoverTree_simple(root):
    if not root:
        return

    nodes = []
    inorder_traversal(root, nodes)

    first = None
    second = None

    for i in range(len(nodes) - 1):
        if nodes[i].val > nodes[i+1].val:
            if first is None:
                first = nodes[i]
            second = nodes[i+1]

    if first and second:
        first.val, second.val = second.val, first.val

def solve(arr):
    root = build_tree(arr)
    recoverTree_simple(root)
    return tree_to_list(root)

if __name__ == "__main__":
    print(solve([1, 3, None, None, 2]))
    print(solve([3, 1, 4, None, None, 2]))

# algorithm time complexity is: O(n)
````

设计一个算法计算从根节点到叶节点生成的所有数字之和 ，分析算法设计思路，计算时间复杂度，并基于python编程实现

e.g.:

 输入：root = [1,2,3], 输出：25

2.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-2.jpg"> </div>

 输入：root = [4,9,0,5,1], 输出：1026

解释：从根到叶子节点路径 4->9->5 代表数字 495，4->9->1 代表数字 491，4->0 代表数字 40。因此和为 495 + 491 + 40 = 1026 。
<div style="text-align:center;width:1000px;"> <img alt="" src="./fig/问题10.jpg"> </div>


idea:

````python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def sumNumbers(root: TreeNode) -> int:
    def dfs(node, current_sum):
        if not node:
            return 0

        current_sum = current_sum * 10 + node.val

        if not node.left and not node.right:
            return current_sum

        return dfs(node.left, current_sum) + dfs(node.right, current_sum)

    return dfs(root, 0)

def build_tree(arr):
    if not arr:
        return None
    root = TreeNode(arr[0])
    queue = [root]
    i = 1
    while queue and i < len(arr):
        node = queue.pop(0)
        if arr[i] is not None:
            node.left = TreeNode(arr[i])
            queue.append(node.left)
        i += 1
        if i < len(arr) and arr[i] is not None:
            node.right = TreeNode(arr[i])
            queue.append(node.right)
        i += 1
    return root

if __name__ == "__main__":
    root1 = build_tree([1, 2, 3])
    print(sumNumbers(root1))

    root2 = build_tree([4, 9, 0, 5, 1])
    print(sumNumbers(root2))
#algorithm time complexity is: O(n)
````
